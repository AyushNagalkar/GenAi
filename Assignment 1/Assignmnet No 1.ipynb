```python
import random
```

```python
filename = "text.txt"
```

```python
# Read text
def read_text(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
    return text.lower().split()
```

```python
# Build Markov Chain
def build_markov_chain(words):
    markov = {}

    for i in range(len(words) - 1):
        current_word = words[i]
        next_word = words[i + 1]

        if current_word not in markov:
            markov[current_word] = []

        markov[current_word].append(next_word)

    return markov
```

```python
def generate_text(markov, start_word, length=20):
    current_word = start_word
    result = [current_word]

    for _ in range(length - 1):
        if current_word not in markov:
            break

        next_word = random.choice(markov[current_word])
        result.append(next_word)
        current_word = next_word

    return " ".join(result)
```

```python

if __name__ == "__main__":
    text = read_text("text.txt")   # put your text here
    markov_chain = build_markov_chain(text)

    start = random.choice(text)
    generated = generate_text(markov_chain, start, 30)

    print("Generated Text:\n")
    print(generated)
```
